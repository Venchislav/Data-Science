```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
# I'm sorry...
df = pd.read_csv('/content/sample_data/penis.csv')
df.head()
```

```python
# 0% condemnation 100% understanding
df.sort_values(by='length_erect', axis=0).head()
```

```python
df = df.drop('Source', axis=1)
df.head()
```

```python
df.Method.unique()
```

```python
df['Method'].mask(df['Method'] == 'Measured', 1.0, inplace=True)
df['Method'].mask(df['Method'] == 'Self reported', 0.0, inplace=True)
```

```python
x = 0
for i in df.Country.unique():
  df.replace(i, float(x), inplace=True)
  x += 1
```

```python
x = 0
for i in df.Region.unique():
  df.replace(i, float(x), inplace=True)
  x += 1
```

```python
df.dtypes
```

```python
X = df.drop('length_erect', axis=1)
y = df.length_erect
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
y_test = np.array(y_test)
```

```python
preds = []
real = []
```

```python
bad = 0
good = 0
for i in range(len(y_pred)):
  if abs(y_test[i] - y_pred[i]) >= 0.95:
    bad += 1
  else:
    good += 1
  preds.append(y_pred[i])
  real.append(y_test[i])


print(f'{good} - good\n {bad} - bad')
```
