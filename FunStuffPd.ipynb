```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/accountancy.csv')
df.head()
```

```python
res = df.groupby(['Executor', 'Type']).aggregate('mean').Salary
```

```python
df_bio = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```
