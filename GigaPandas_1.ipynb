```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/movie_metadata.csv')
df.head()
```

```python
genres = df[['movie_title', 'genres']]
genres.head()
```

```python
for row in genres.head().values:
  for column in row:
    print(column)
```

```python
# fun thing we can do:

%%timeit
def reversator(value):
  return value[::-1]

for row in genres.head(10).values:
  for col in row:
    print(reversator(col))
```

```python
for i, row in genres.head(10).iterrows():
  print(row.map(reversator))
```

```python
for i, col in genres.iteritems():
  print(col)
```

```python
df = pd.DataFrame(range(10000000))
```

```python
import pandas as pd
import numpy as np
from time import time

before = time()
df.apply(np.mean)   #подставляете сюда элементы из списка, потом располагаете элементы по возрастанию в ответе
after = time()
print(after - before)
```

```python
before = time()
df.mean(axis=0)   #подставляете сюда элементы из списка, потом располагаете элементы по возрастанию в ответе
after = time()
print(after - before)
```

```python
before = time()
df.apply('mean')   #подставляете сюда элементы из списка, потом располагаете элементы по возрастанию в ответе
after = time()
print(after - before)
```
