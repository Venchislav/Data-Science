```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sp = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
sp.head()
```

```python
sp = sp.rename(columns={'parental level of education': 'parental level of education'.replace(' ', '_'),
                        'test preparation course': 'test preparation course'.replace(' ', '_'),
                        'math score': 'math score'.replace(' ', '_'),
                        'reading score': 'reading score'.replace(' ', '_'),
                        'writing score': 'writing score'.replace(' ', '_')})
```

```python
sp.head()
```

```python
sp.math_score.hist()
```

```python
sp.plot.scatter(x='math_score', y='reading_score')
```

```python
sns.lmplot(x='math_score', y='reading_score', data=sp)
```

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=sp)
```

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', fit_reg=False, data=sp)
```

```python
iris = pd.read_csv('https://stepik.org/media/attachments/course/4852/iris.csv')
iris.head()
```

```python
iris = iris.drop(['Unnamed: 0'], axis=1)
```

```python
for col in iris:
  sns.kdeplot(data=iris[col])
plt.xlabel('characteristics')
plt.legend([col for col in iris])
```

```python
sns.violinplot(y=iris["petal length"])
```

```python
sns.pairplot(iris)
```
