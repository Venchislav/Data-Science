```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
df.head()
```

```python
df = df.rename(columns={
    'parental level of education': 'parental level of education'.replace(' ', '_'),
    'test preparation course': 'test preparation course'.replace(' ', '_'),
    'math score': 'math score'.replace(' ', '_'),
    'reading score': 'reading score'.replace(' ', '_'),
    'writing score': 'writing score'.replace(' ', '_')

})
```

```python
df.groupby('gender').mean()
```

```python
df.groupby('gender', as_index=False).aggregate({'math_score': 'mean', 'writing_score': 'mean'})\
  .rename(columns={'math_score': 'math_score_mean',
                   'writing_score': 'writing_score_mean'})
```

```python
grouped_df = df.groupby(['gender', 'race/ethnicity'], as_index=False).aggregate({'math_score': 'mean', 'writing_score': 'mean'})\
  .rename(columns={'math_score': 'math_score_mean',
                   'writing_score': 'writing_score_mean'})
```

```python
grouped_df
```

```python
a = grouped_df.sort_values(['gender', 'math_score_mean'])
a
```

```python
b = grouped_df.sort_values(['gender', 'writing_score_mean'])
b
```

```python
# add total score column
df['total_score'] = df.filter(like='score').sum(axis=1)
```

```python
df
```

```python
# add column with log
df = df.assign(total_score_log=np.log(df.total_score))
df
```

```python
# delete column
df.drop('total_score', axis=1)
```
