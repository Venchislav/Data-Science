```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/invasion.csv')
df.head()
```

```python
df.shape
```

```python
X_train = df.drop('class', axis=1)
y_train = df['class']
```

```python
transport = {'transport' :  0,  'fighter' :  1,  'cruiser' : 2}
y_train = df['class'].map(transport)
```

```python
clf = RandomForestClassifier()

params = {'n_estimators': range(10, 50, 10),
          'max_depth': range(1, 12, 2),
          'min_samples_leaf': range(1, 7),
          'min_samples_split': range(2, 9, 2)}

grid = GridSearchCV(clf, params, cv=3)
grid.fit(X_train, y_train)
```

```python
best_clf = grid.best_estimator_
best_clf.fit(X_train, y_train)
```

```python
df_test = pd.read_csv('https://stepik.org/media/attachments/course/4852/operative_information.csv')
df_test.head()
```

```python
y_pred = best_clf.predict(df_test)
```

```python
pd.Series(y_pred).value_counts()
```

```python
features = pd.DataFrame(best_clf.feature_importances_, index=X_train.columns, columns=['importance'])
features.sort_values(by='importance', ascending=False)
```
