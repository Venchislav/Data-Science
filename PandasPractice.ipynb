```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat_1.csv')
df.head(10)
```

```python
df[df['n_users'] > 0]['n_users'].median()
```

```python
df['session_value'] = df['session_value'].fillna(0)
# df.query('n_users < 0').aggregate('median')
df['n_users'] = df['n_users'].map(lambda x: df[df['n_users'] >= 0]['n_users'].median() if x < 0 else x)
df.head(10)
```

```python
mean_session_value_data = df.groupby('group').agg({'session_value': 'mean'})
mean_session_value_data = mean_session_value_data.rename(columns={'session_value': 'mean_session_value'})
mean_session_value_data.head()
```
