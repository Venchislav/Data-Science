```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Now we will be working with NaNs...
```

```python
df = pd.read_csv('/content/sample_data/train.csv')
```

```python
df.head()
```

```python
import missingno as msno
```

```python
ax = msno.matrix(df.sample(500))
ax.get_figure().savefig('/content/sample_data/bro.png')
```

```python
# It's not a barcode. It's a NaNs values concentraition
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
(1 - df.isnull().mean()).abs().plot.bar(ax=ax)
fig.savefig('/content/sample_data/bro.png')
```

```python
# but there's a smater way to impute NaNs then fillna
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Here we choose columns with number datatype
num_cols = df.select_dtypes(include="number").columns
```

```python
# and we just impute it with means
# it also takes additional strategy arg with 'median' or 'most_frequent' value
im = SimpleImputer()
imputed = im.fit_transform(df[num_cols])
```
